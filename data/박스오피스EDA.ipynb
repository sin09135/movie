## 박스오피스 EDA

```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/boxoffice_15_23_total.csv')
```

```python
df.info()
```

### 데이터 재구성

```python
agg_df = df.groupby('movieNm').agg({
    'movieCd': 'first', # 영화 구분 코드
    'openDt': 'first', # 개봉일
    'salesAcc': 'max', # 누적 매출
    'audiAcc': 'max', # 누적 매출
    'scrnCnt': 'sum', # 스크린 수 합산
    'typeNm' : 'first',
    'nationNm':'first',
    'genreNm':'first',
    'director':'first',
    'actor':'first',
    'showTypeGroupNm':'first',
    'showTypeNm':'first',
    'audits':'first'
    
}).reset_index()
```
