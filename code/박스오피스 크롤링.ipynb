```python
import requests
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

def boxoffice_requests_start_to_end(start, end,key):
    # 시작일과 종료일 datetime으로 변환
    start_date = datetime.strptime(start, "%Y-%m-%d")
    end_date = datetime.strptime(end, "%Y-%m-%d")

    # 기본 URL
    url_base = 'http://kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json'

    # append할 리스트 생성
    daily_boxoffice_data = []

    # 시작일, 종료일까지 호출
    current_date = start_date
    while current_date <= end_date:
        targetDt = current_date.strftime("%Y%m%d")
        url = f'{url_base}?key={key}&targetDt={targetDt}'
        response = requests.get(url)
        r_data = response.json()

        # 해당 날짜에 박스오피스 데이터가 있는 경우 > 추가
        if r_data['boxOfficeResult']['dailyBoxOfficeList']:
            df = pd.DataFrame(r_data['boxOfficeResult']['dailyBoxOfficeList'])
            # 박스오피스 기준 날짜를 컬럼에 추가 > 추후 파생변수 생성을 위함
            df['current_date'] = current_date.strftime("%Y-%m-%d")
            daily_boxoffice_data.append(df)

        # 다음 날짜로 이동
        current_date += timedelta(days=1)

    # 모든 데이터를 하나의 DataFrame으로 결합
    daily_boxoffice_df = pd.concat(daily_boxoffice_data, ignore_index=True)
    daily_boxoffice_df.to_csv('boxoffice_15_19.csv', index = False, encoding = 'utf-8-sig')
    
    
    return daily_boxoffice_df
```

```python
import requests
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

def boxoffice_requests_start_to_end(start, end, key, itemPerPage="10", multiMovieYn=None, repNationCd=None, wideAreaCd=None, title="boxoffice"):
    # 시작일과 종료일 datetime으로 변환
    start_date = datetime.strptime(start, "%Y-%m-%d")
    end_date = datetime.strptime(end, "%Y-%m-%d")

    # 기본 URL
    url_base = 'http://kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json'

    # append할 리스트 생성
    daily_boxoffice_data = []

    # 시작일, 종료일까지 호출
    current_date = start_date
    while current_date <= end_date:
        targetDt = current_date.strftime("%Y%m%d")
        url = f'{url_base}?key={key}&targetDt={targetDt}&itemPerPage={itemPerPage}'
        
        # multiMovieYn이나 repNationCd, wideAreaCd가 주어진 경우에만 추가
        if multiMovieYn:
            url += f'&multiMovieYn={multiMovieYn}'
        if repNationCd:
            url += f'&repNationCd={repNationCd}'
        if wideAreaCd:
            url += f'&wideAreaCd={wideAreaCd}'
        
        response = requests.get(url)
        r_data = response.json()

        # 해당 날짜에 박스오피스 데이터가 있는 경우 > 추가
        if r_data['boxOfficeResult']['dailyBoxOfficeList']:
            df = pd.DataFrame(r_data['boxOfficeResult']['dailyBoxOfficeList'])
            # 박스오피스 기준 날짜를 컬럼에 추가 > 추후 파생변수 생성을 위함
            df['current_date'] = current_date.strftime("%Y-%m-%d")
            daily_boxoffice_data.append(df)

        # 다음 날짜로 이동
        current_date += timedelta(days=1)

    # 모든 데이터를 하나의 DataFrame으로 결합
    daily_boxoffice_df = pd.concat(daily_boxoffice_data, ignore_index=True)
    daily_boxoffice_df.to_csv(f'../data/{title}.csv', index=False, encoding='utf-8-sig')
    
    return daily_boxoffice_df
```
