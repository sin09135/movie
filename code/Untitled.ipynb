```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 주어진 데이터
data = {
    'Feature': ['정체기 (s)', '팬데믹 (p)', '회복기 (r)'],
    '1_weeks': [0.79306, 0.85838, 0.64932],
    '2_weeks': [0.90308, 0.96374, 0.89435],
    '3_weeks': [0.86076, 0.92066, 0.92722],
    '4_weeks': [0.82327, 0.63285, 0.89577],
    '5_weeks': [0.71696, 0.54179, 0.86303],
    '6_weeks': [0.62352, 0.56753, 0.79534],
    '7_weeks': [0.45686, 0.47344, 0.56460],
    '8_weeks': [0.38998, 0.26477, 0.42787]
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# Feature 열을 인덱스로 설정
df.set_index('Feature', inplace=True)

# 히트맵 그리기
plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('주차별 정체기, 팬데믹, 회복기 feature와 최종관객수 상관계수')
plt.xlabel('주차')
plt.ylabel('Feature')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Given data
data = {
    'Feature': ['Stagnation (s)', 'Pandemic (p)', 'Recovery (r)'],
    '1_weeks': [0.79306, 0.85838, 0.64932],
    '2_weeks': [0.90308, 0.96374, 0.89435],
    '3_weeks': [0.86076, 0.92066, 0.92722],
    '4_weeks': [0.82327, 0.63285, 0.89577],
    '5_weeks': [0.71696, 0.54179, 0.86303],
    '6_weeks': [0.62352, 0.56753, 0.79534],
    '7_weeks': [0.45686, 0.47344, 0.56460],
    '8_weeks': [0.38998, 0.26477, 0.42787]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set the 'Feature' column as index
df.set_index('Feature', inplace=True)

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Coefficients between Stagnation, Pandemic, Recovery Features and Final Audience by Week')
plt.xlabel('Week')
plt.ylabel('Feature')
plt.show()
```
